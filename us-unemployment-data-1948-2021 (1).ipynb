# * Analyze and Explore DataSet üòÉ‚ù§Ô∏èü§ù


```python
#Importing the basic librarires fot analysis

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("ggplot")  #using style ggplot

%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import datetime as dt
import plotly.graph_objects as go
import plotly.express as px
```

```python
#Importing the dataset
df =pd.read_csv("../input/unemployment-data-19482021/unemployment_rate_data.csv")


# look the data set
df.head()
```

```python
# looking the shape DataSet
df.shape
```

```python
#Checking the dtypes of all the columns

df.info()
```

```python
#Conversion data type column - Date from object to Datetime

df["date"]=pd.to_datetime(df["date"])
```

```python
#checking null value 
df.isna().sum()
```

- No missing value

```python
# look  describe data set
df.describe().round(2)
```

# Some Visualizations üòÉ‚ù§Ô∏èü§ù

```python
# show the relation years to general Unrated
plt.figure(figsize=(15,7))
sns.lineplot(data=df , x="date",y="unrate")
```

- In general, the unemployment rate is between 3-8 - during these years, but a high rise in 2020 is due to Corona

```python
# show the relation years to Unrated men and women - interactive graph

px.scatter_3d(df,x='date',y="unrate_men",z="unrate_women",color="unrate_women")
```

```python
#  show the relation years to Unrated men and women also general Unrated
 
plt.figure(figsize=(30,10))
sns.lineplot(data=df,x="date",y="unrate",label="un rate")
sns.lineplot(data=df,x="date",y="unrate_men",label="Men")
sns.lineplot(data=df,x="date",y="unrate_women",label="Women")


plt.legend
```

- The unemployment rate is about the same for men and women

```python
# show the Unemployment category

plt.figure(figsize=(30,10))
sns.lineplot(data=df,x="date",y="unrate_16_to_17",label="unrate_16_to_17")
sns.lineplot(data=df,x="date",y="unrate_18_to_19",label="unrate_18_to_19")
sns.lineplot(data=df,x="date",y="unrate_20_to_24",label="unrate_20_to_24")
sns.lineplot(data=df,x="date",y="unrate_25_to_34",label="unrate_25_to_34")
sns.lineplot(data=df,x="date",y="unrate_35_to_44",label="unrate_35_to_44")
sns.lineplot(data=df,x="date",y="unrate_45_to_54",label="unrate_45_to_54")
sns.lineplot(data=df,x="date",y="unrate_55_over",label="unrate_55_over")

plt.legend
```

- As the age increases, unemployment decreases, and this is due to experience

```python
sns.pairplot(df)
```

```python
#add new column Year for makeing group by 
df['Year']= df['date'].dt.year
avg_year=df.groupby("Year")["unrate","unrate_men","unrate_women"].mean().round(2)
avg_year
```

```python
# Visualizations group by
avg_year.plot(figsize=(20,7))
```

```python
avg_cat=df.groupby("Year")["unrate_16_to_17","unrate_18_to_19","unrate_20_to_24","unrate_25_to_34","unrate_35_to_44","unrate_45_to_54","unrate_55_over"].mean().round(2)
avg_cat
```

```python
# Visualizations group by
avg_cat.plot(figsize=(20,7))
```
